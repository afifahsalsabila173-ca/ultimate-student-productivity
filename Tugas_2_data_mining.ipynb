```python
df_can = pd.read_csv(
'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv'
)

df_can.set_index('Country', inplace=True)

years = list(map(str, range(1980,2014)))
x = years
y = df_can.loc['Indonesia', years]

plt.figure(figsize=(10,5))

plt.plot(x,y,'ro-')  # merah titik + garis
plt.title("Tren Imigrasi dari Indonesia ke Kanada")
plt.xlabel("Tahun")
plt.ylabel("Jumlah Imigran")

plt.show()
```

menampilkan grafik garis yang menunjukkan perubahan (tren) jumlah imigran dari Indonesia ke Kanada setiap tahun dari 1980 sampai 2013.

Data diambil dari dataset imigrasi Kanada, kemudian dipilih khusus data Indonesia, dan divisualisasikan menggunakan grafik garis dengan titik merah dan garis penghubung ('ro-') agar perubahan jumlah imigran tiap tahun terlihat jelas.

```python
plt.figure(figsize=(10,5))

plt.plot(df_can.loc['Indonesia', years], 'ro', label="Indonesia")
plt.plot(df_can.loc['China', years], 'g*', label="China")
plt.plot(df_can.loc['India', years], 'bs', label="India")
plt.plot(df_can.loc['Philippines', years], 'g^', label="Philippines")

plt.title("Perbandingan Imigrasi Beberapa Negara")
plt.xlabel("Tahun")
plt.ylabel("Jumlah Imigran")

plt.legend()
plt.show()
```

membuat grafik yang membandingkan tren jumlah imigran dari beberapa negara (Indonesia, China, India, dan Philippines) ke Kanada dari tahun 1980–2013. Setiap negara ditampilkan dengan warna dan bentuk marker yang berbeda agar mudah dibedakan, dan plt.legend() digunakan untuk menampilkan keterangan nama negara pada grafik.

```python
x = list(range(0,34))
y = df_can.loc['Indonesia', years]

plt.plot(x,y,'b-')

plt.axis([0,34,0,2000])
# xmin xmax ymin ymax

plt.title("Axis Setting Imigrasi Indonesia")

plt.show()
```

menampilkan grafik garis jumlah imigran dari Indonesia ke Kanada dengan pengaturan batas sumbu (axis). Perintah plt.axis([0,34,0,2000]) digunakan untuk menentukan batas minimum dan maksimum pada sumbu X dan Y, yaitu sumbu X dari 0–34 (jumlah tahun) dan sumbu Y dari 0–2000 (jumlah imigran).

```python
top10 = df_can.sort_values(by='Total', ascending=False).head(10)

top10['Total'].plot(kind='bar', figsize=(10,6), color='green')

plt.title("10 Negara dengan Imigrasi Terbesar ke Kanada")
plt.xlabel("Negara")
plt.ylabel("Jumlah Imigran")

plt.show()
```

menampilkan grafik batang yang menunjukkan 10 negara dengan jumlah imigran terbesar ke Kanada. Data terlebih dahulu diurutkan berdasarkan total imigrasi dari terbesar ke terkecil, kemudian diambil 10 negara teratas dan divisualisasikan dalam grafik batang berwarna hijau untuk memudahkan perbandingan jumlah imigran antar negara.

```python
top5 = df_can.sort_values(by='Total', ascending=False).head(5)

plt.figure(figsize=(8,8))

plt.pie(top5['Total'],
labels=top5.index,
autopct='%1.1f%%',
startangle=90)

plt.title("Persentase Imigrasi 5 Negara Terbesar ke Kanada")

plt.show()
```

membuat diagram lingkaran (pie chart) yang menampilkan persentase kontribusi imigran dari 5 negara dengan jumlah imigrasi terbesar ke Kanada. Data diurutkan terlebih dahulu berdasarkan total imigrasi, kemudian ditampilkan dalam bentuk persentase menggunakan autopct='%1.1f%%' agar terlihat bagian kontribusi tiap negara.

```python
quota = 500000

proporsi = top5['Total']/top5['Total'].sum()

kuota_negara = proporsi*quota

print(kuota_negara)
```

menghitung pembagian kuota imigrasi sebesar 500.000 orang kepada 5 negara teratas secara proporsional berdasarkan total imigrasi sebelumnya. Proporsi tiap negara dihitung dari total imigrasinya, lalu dikalikan dengan kuota untuk mendapatkan jumlah kuota yang diterima masing-masing negara.

```python
import matplotlib.pyplot as plt

indonesia = df_can.loc['Indonesia', years]
india = df_can.loc['India', years]

plt.figure(figsize=(8,6))

plt.scatter(indonesia, india, color='purple', marker='o')

plt.title("Hubungan Imigrasi Indonesia dan India")
plt.xlabel("Imigran Indonesia")
plt.ylabel("Imigran India")

plt.show()
```

membuat scatter plot yang menunjukkan hubungan atau perbandingan jumlah imigran dari Indonesia dan India ke Kanada. Setiap titik pada grafik mewakili pasangan jumlah imigran Indonesia dan India pada tahun yang sama sehingga dapat dilihat pola atau korelasinya.

menampilkan grafik batang bertumpuk (stacked bar chart) yang menunjukkan jumlah imigrasi dari 5 negara dengan total imigrasi terbesar ke Kanada setiap tahun. Grafik ini membantu melihat kontribusi masing-masing negara serta total imigrasi per tahun dalam satu tampilan.

```python
negara = df_can.loc[['Indonesia','India','China','Philippines'], years].T

negara.plot(kind='box', figsize=(8,6))

plt.title("Distribusi Imigrasi Beberapa Negara")
plt.ylabel("Jumlah Imigran")

plt.show()
```

menampilkan box plot yang menunjukkan distribusi jumlah imigrasi dari Indonesia, India, China, dan Philippines ke Kanada selama beberapa tahun. Grafik ini membantu melihat persebaran data, nilai median, serta kemungkinan adanya nilai yang sangat tinggi atau rendah (outlier) pada masing-masing negara.

Boxplot menunjukkan median, quartile, dan outlier.
garis tengah → median
kotak → rentang data utama
titik → outlier

```python
top5 = df_can.sort_values(by='Total', ascending=False).head(5)

explode = [0.1,0,0,0,0]

plt.figure(figsize=(8,8))

plt.pie(
top5['Total'],
labels=top5.index,
autopct='%1.1f%%',
startangle=90,
explode=explode
)

plt.title("Persentase Imigrasi 5 Negara Terbesar")

plt.show()
```
